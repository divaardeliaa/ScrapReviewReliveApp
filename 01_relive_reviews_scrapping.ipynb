```python
from google_play_scraper import app
import pandas as pd
import numpy as np
```

##Scraping Relive App Review by its id

```python
#Android App relive from Google Play at https://play.google.com/store/apps/details?id=cc.relive.reliveapp&pcampaignid=web_share
#The apps ID found in the link after id=cc.relive.reliveapp&pcampaignid=web_share

from google_play_scraper import app, Sort, reviews_all

relive_reviews = reviews_all(
    'cc.relive.reliveapp',
    sleep_milliseconds=0, # defaults to 0
    lang='en', # defaults to 'en'
    sort=Sort.NEWEST, # defaults to Sort.MOST_RELEVANT
)
```

##Saving results into a dataframe

```python
#Save relive reviews into dataframe named df_reliverev
df_reliverev = pd.DataFrame(np.array(relive_reviews),columns=['content'])
df_reliverev = df_reliverev.join(pd.DataFrame(df_reliverev.pop('content').tolist()))
df_reliverev.to_csv(r'df_reliverev', index=False)
```
